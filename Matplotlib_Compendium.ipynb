### Line styles 
The next code example was taken from https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
linestyle_str = [
     ('solid', 'solid'),      # Same as (0, ()) or '-'
     ('dotted', 'dotted'),    # Same as (0, (1, 1)) or '.'
     ('dashed', 'dashed'),    # Same as '--'
     ('dashdot', 'dashdot')]  # Same as '-.'

linestyle_tuple = [
     ('loosely dotted',        (0, (1, 10))),
     ('dotted',                (0, (1, 1))),
     ('densely dotted',        (0, (1, 1))),

     ('loosely dashed',        (0, (5, 10))),
     ('dashed',                (0, (5, 5))),
     ('densely dashed',        (0, (5, 1))),

     ('loosely dashdotted',    (0, (3, 10, 1, 10))),
     ('dashdotted',            (0, (3, 5, 1, 5))),
     ('densely dashdotted',    (0, (3, 1, 1, 1))),

     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))]


def plot_linestyles(ax, linestyles):
    X, Y = np.linspace(0, 100, 10), np.zeros(10)
    yticklabels = []

    for i, (name, linestyle) in enumerate(linestyles):
        ax.plot(X, Y+i, linestyle=linestyle, linewidth=1.5, color='black')
        yticklabels.append(name)

    ax.set(xticks=[], ylim=(-0.5, len(linestyles)-0.5),
           yticks=np.arange(len(linestyles)), yticklabels=yticklabels)

    # For each line style, add a text annotation with a small offset from
    # the reference point (0 in Axes coords, y tick value in Data coords).
    for i, (name, linestyle) in enumerate(linestyles):
        ax.annotate(repr(linestyle),
                    xy=(0.0, i), xycoords=ax.get_yaxis_transform(),
                    xytext=(-6, -12), textcoords='offset points',
                    color="blue", fontsize=8, ha="right", family="monospace")


fig, (ax0, ax1) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [1, 3]},
                               figsize=(10, 8))

plot_linestyles(ax0, linestyle_str[::-1])
plot_linestyles(ax1, linestyle_tuple[::-1])

plt.tight_layout()
plt.show()
```

## Pie Charts <a class="anchor" id="pie"></a>

You can find all the parameters at: *https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html*

```python
import random 
fig_7 = plt.figure(figsize=(8,5), dpi=100)
axes_7 = fig_7.add_axes([0.1,0.1,0.9,0.9])

#Create a pie chart
types = ['Opction1', 'Opction2', 'Opction3', 'Opction4', 'Opction5']
percent = [6,28,34,8,24]

color = []
for i in range(5):
    rgb = (random.uniform(0,.5), random.uniform(0,.5), random.uniform(0,.5))
    color.append(rgb)
    
explode = [0]*5
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(percent, explode=explode, labels=types,
                                   colors=color, autopct='%1.0f%%', shadow=True,
                                   startangle=140, textprops=dict(color='w'))

plt.legend(wedges, types, loc='right', bbox_to_anchor=(1, 0, 0.5, 1))
```

## Quiver Plots <a class="anchor" id="quiver"></a> 
Is a type of 2D plot which shows vector lines as arrows. The sintax is:

*ax.quiver(x_pos, y_pos, x_dir, y_dir, color)*

Here **x_pos** and **y_pos** are the starting positions of the arrow while **x_dir** and **y_dir** are the directions of the arrow.

- For example, we graph an arrow that goes from (0,0) to (1,1),

```python
X = 0
Y = 0
U = 1
V = 1
plt.quiver(X,Y,U,V)
```

Now, we will use the function that returns a grid of points to be the start of our arrows. So you don't have to draw one by one.

- **np.mgrid**: this function returns mesh-grid ndarrays all of the same dimensions
- **meshgrid**: return coordinate matrices from coordinate vectors

```python
n = 20
X, Y = np.mgrid[0:n, 0:n]
plt.scatter(X, Y)
plt.show()
```

```python
So, this is the grid of our initial points of the arrows
```

```python
n = 20
X, Y = np.mgrid[0:n, 0:n]
U = np.sin(X)
V = np.sin(Y)
plt.quiver(X, Y, U, V)
```

You can choose that the initial points of your mesh correspond to a certain equation

```python
x = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)
y = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)

X,Y = np.meshgrid(x,y)

U = np.sin(X)*np.cos(Y)
V = -np.cos(X)*np.sin(Y)

fig, ax = plt.subplots(figsize=(7,7))

ax.quiver(X,Y,U,V)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([0,2*np.pi,0,2*np.pi])
ax.set_aspect('equal')

plt.show()
```

```python
n = 8 
X, Y = np.mgrid[0:n, 0:n]
T = np.arctan2(Y - n / 2., X - n/2.)
R = 10 + np.sqrt((Y - n / 2.0) ** 2 + (X - n / 2.0) ** 2)
U, V = R * np.cos(T), R * np.sin(T)
plt.axes([0.025, 0.025, 0.95, 0.95])
plt.quiver(X, Y, U, V, R, alpha=.5) #Remember that alpha define the transparency
plt.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=.5)
plt.xlim(-1, n)
plt.xticks(())
plt.ylim(-1, n)
plt.yticks(())
plt.show()
```

## Polar Plots<a class="anchor" id="polar"></a> 
A point in polar coordinates is represented as (r, theta). Here, r is its distance from the origin and theta is the angle at which r has to be measured from origin.

The sintax is: **plt.polar(theta, r, **kwargs)** 

Let's plots some polar functions.
- **Example 1**: A circle is a round plane figure whose boundary (the circumference) consists of points equidistant from a fixed point (the center). So the distance r is constant.

```python
# setting the axes projection as polar
plt.axes(projection = 'polar')
  
# setting the radius
r = 2
  
# creating an array containing the
# radian values
rads = np.arange(0, (2 * np.pi), 0.01)
  
# plotting the circle
for rad in rads:
    plt.polar(rad, r, 'r.')
  
# display the Polar plot
plt.show()
```

Otherwise the circle where its r does not depend on the angle, in many polar graphs it will depend and for this the following example will help you

```python
# setting the axes projection as polar
plt.axes(projection = 'polar')
  
# creating an array containing the radian values
rads = np.arange(0, (2 * np.pi), 0.01)
  
# plotting 
for rad in rads:
    r = 2-2*np.sin(rad) # setting the radius
    plt.polar(rad, r, 'b.')
  
# display the Polar plot
plt.show()
```

## 3D Surface <a class="anchor" id="3D"></a>

```python
# Needed for creating 3D axes
from mpl_toolkits import mplot3d

fig_9 = plt.figure(figsize=(8,5),dpi=100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection='3d')

# Create a 3D scatterplot
# Te darker points are represented that way to seem closer to you
z_3 = 15 * np.random.random(100)
x_3 = np.sin(z_3) * np.random.randn(100)
y_3 = np.cos(z_3) * np.random.randn(100)
# axes_9.scatter3D(x_3, y_3, z_3, c=z_3, cmap='Blues')

# You can create contour plots by defining a function for z based on x & y
def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x_4 = np.linspace(-6, 6, 30)
y_4 = np.linspace(-6, 6, 30)

# Creates a rectangular grid out of 2 given 1D arrays
x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4,y_4)

# Change viewing angle to reorient camera 60 degrees and rotate 55
axes_9.view_init(45,55)

# Provide x, y, z, contours and color map
# axes_9.contour3D(x_4,y_4,z_4,80,cmap='Blues')
axes_9.set_xlabel('x')
axes_9.set_ylabel('y')
axes_9.set_zlabel('z')

# You can create wireframes
# axes_9.plot_wireframe(x_4,y_4,z_4,color='blue')

# You can create surface plots which is wireframe with filled faces
axes_9.plot_surface(x_4,y_4,z_4, rstride=1, cstride=1,cmap='Blues',edgecolor='none')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2, 2)
plt.xticks(())
plt.yticks(())
ax.set_zticks(())

ax.text2D(0.05, .93, " 3D plots \n",
    horizontalalignment='left',
    verticalalignment='top',
    size='xx-large',
    bbox=dict(facecolor='white', alpha=1.0),
    transform=plt.gca().transAxes)

ax.text2D(0.05, .87, " Plot 2D or 3D data",
    horizontalalignment='left',
    verticalalignment='top',
    size='large',
    transform=plt.gca().transAxes)

plt.show()
```

- Plot an ellipse with major axis 12 and minor axis 6.

```python
# setting the axes
# projection as polar
plt.axes(projection = 'polar')
  
# setting the values of
a = 12 # semi-major
b = 8  # semi-minor
   
# creating an array containing the radian values
rads = np.arange(0, (2 * np.pi), 0.01)
  
# plotting the ellipse
for rad in rads:
    r = (a*b)/np.sqrt((a*np.sin(rad))**2 + (b*np.cos(rad))**2)
    plt.polar(rad, r, 'g.')
  
# display the polar plot
plt.show()
```
